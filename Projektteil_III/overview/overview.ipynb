```python
# Import needed libaries
import numpy as np
import pandas as pd
```

```python
# Read DataFrame
df4 = pd.read_pickle('df4.pkl')
```

```python
# Select features
feats = ['p_product_group','p_product_type','p_product','qc_minimalersalzrueckhalt','qc_salzrckhalt']
#indexes = ['p_product_group','p_product_type','p_product']
```

```python
# Select featurs opon for "split-apply" evaluation
groupCols = ['p_product_group','p_product_type','p_product']
```

```python
# Calculates overview
d = (df4[feats]
 .assign(lows = df4['qc_salzrckhalt']<df4['qc_minimalersalzrueckhalt'])
 .groupby(groupCols)
 .agg({
     'qc_minimalersalzrueckhalt':['mean'],
     'qc_salzrckhalt':['mean','size'],
     'lows':['sum']
 })
)
# Calculates relative count of observations with lower Salt Retention than the Specification
d[('lows','freq')] = d[('lows','sum')]/d.loc[:,('qc_salzrckhalt','size')]
d
```
