# 1. Introduction

## 1.1. Reverse osmosis membrane elements Lewabrane®
All Lewabrane® membrane elements consist of spirally wound thin-film composite membranes that have been specially developed for the highly efficient treatment of drinking and process water. The following figure shows the most important main components and the dimensions of an 8'' membrane element (source: Lewabrane® brochure / Lanxess) 

<img src=element_dimensions.png width=600>


| <i></i>       | A (length)     | B (diameter)  |C (ID)         |
|:------------- |:--------------:|--------------:|--------------:|
| Metric Units  | 1016 mm        |201 mm         |29 mm          |
| US units      | 40 in          |7.9 in         |1.125 in       |

Due to their special chemistry, the elements are characterized by their particularly high retention capacity for critical substances such as borates, nitrates and silicic acid. For this reason, these membrane elements are mainly used for the desalination of seawater, brackish water and slightly salty water. The most relevant applications in water treatment are listed below (source: Lewabrane® brochure / Lanxess)

- Seawater desalination in municipal and industrial plants
- Production of boiler feed water for water-steam circuits
- Complete desalination and particle removal of process water for the production of ultrapure water
- Decentralized water desalination, for example for car washes, laundries or on ships
- Groundwater remediation and recharge for the production of drinking water
- Treatment of industrial and municipal waste water

## 1.2. Original data set and membrane types

The original data set comes from quality control in the fully automated Lewabrane® production plant in Bitterfeld. After a close examination of the columns, it can be seen that each row of the data set reflects the results of a quality control. The fact that the value of the roll_position column for a particular roll always increases indicates that the product properties are checked several times for the same roll. The roll_position column is therefore the exact position where each quality control is performed. Of course, each column also contains information about the previous manufacturing steps.

In order to simplify later data cleaning and exploration, it is first of all necessary to identify the different membrane types in the data set. Using the columns p_product_group, p_product_type, p_product and p_product_full_name, the roles in the data set can be assigned to a commercial product by comparing the properties of the columns qc_minimum salt retention and qc_minimum flow rate with the properties listed in the Lewabrane® brochure.

# 5. Group Columns

In this section we transform the dataset in one with a multiindex for the columns. 

## 5.1 Sort the columns by process step

* Process step 1: Coating with ps
* Process step 2: Coating with pa
* Process step 3: Assembling
* Process step 4: Quality control

```python
col_names=list(df9)
```

## 5.2 Define the process steps

```python
process_steps=['1ps','2pa','3ass', '4qc'] #Numbers to arrange process steps in the correct order
```

```python
order_by_process_dict={}
```

## 5.3 Grouping columns using a dictionary

```python
for col in col_names:
    
    
    #Prozessschritt 1
    if col.startswith('ps'):
        
        order_by_process_dict[col]=process_steps[0]
    
    #Prozessschritt 2
    elif col.startswith('pa_') or col.startswith('ct') or col.startswith('pp') or col.startswith('qt') or col.startswith('reaction') or col.endswith('median'):
        
        order_by_process_dict[col]=process_steps[1]
        
    elif col.startswith('wind') or col.startswith('sc')or col.startswith('leak') or col.startswith('assembl'):
        
        order_by_process_dict[col]=process_steps[2]
        
    elif col.startswith('qc_') or col.startswith('element') or col.startswith('sum') or col.startswith('roll') or col.startswith('total') or col.startswith('p_') or col.startswith('nr') or col.startswith('component') or col.startswith('derived'):
        
        order_by_process_dict[col]=process_steps[3]
```

```python
df10=pd.concat(dict((*df9.groupby(order_by_process_dict, 1),)), axis=1)
df10.columns.set_levels(['ps','pa','ass','qc'],level=0,inplace=True)
```
